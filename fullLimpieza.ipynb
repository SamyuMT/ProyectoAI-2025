### Importamos la libreria propia

```python
from libreria import *
from pathlib import Path

RAW_DIR = Path("./data/raw")      # <‑‑ coloca los ZIP/CSV descargados aquí
PREP_DIR = Path("./data/processed")
PREP_DIR.mkdir(parents=True, exist_ok=True)
```

### Hallamos los paths de los datasets

```python
DATASETS = {
    "II-2024": RAW_DIR / "CEEDEST_2024_T2.csv",
    "III-2024": RAW_DIR / "CEEDEST_2024_T3.csv",
    "IV-2024": RAW_DIR / "CEEDEST_2024_T4.csv",
}
```

```python
df1 = pl.read_csv(
    DATASETS["II-2024"],
    infer_schema_length=None,
    low_memory=True
)
df2 = pl.read_csv(
    DATASETS["III-2024"],
    infer_schema_length=None,
    low_memory=True
)
df3 = pl.read_csv(
    DATASETS["IV-2024"],
    infer_schema_length=None,
    low_memory=True
)
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
df1_cleaned = df_full_clean(df1)
```

```python
df2_cleaned = df_full_clean(df2)
```

```python
df3_cleaned = df_full_clean(df3)
```

```python
df1_cleaned.write_csv(PREP_DIR / "dataset_II_Clean.csv")
df2_cleaned.write_csv(PREP_DIR / "dataset_III_Clean.csv")
df3_cleaned.write_csv(PREP_DIR / "dataset_IV_Clean.csv")
```

```python
TARGET = "PRECIOVTAX"          # ← cambia aquí si tu objetivo se llama distinto
corr_target(TARGET, df1_cleaned)
corr_target(TARGET, df2_cleaned)
corr_target(TARGET, df3_cleaned)
```

```python
dffull = pl.concat([df1_cleaned, df2_cleaned, df3_cleaned])
```

```python
dffull.shape
```

```python
for col, nulos in val_nulos(dffull).items():
    print(f"{col}: {nulos}")
```
